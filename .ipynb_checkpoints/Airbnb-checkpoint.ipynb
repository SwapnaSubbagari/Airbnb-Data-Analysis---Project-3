```python
#import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import CSV data
csv = 'Resources/AB_US_2020.csv'
data = pd.read_csv(csv)
data.head()
```

```python
#drop unneeded columns
newdf = data.drop(['name','last_review'], axis=1)
newdf.head()
```

```python
#Identify data characteristics
newdf.info()
```

```python
#look at the missing data
newdf.isna().sum()
```

```python
#??view a heatmap of the missing data using seaborn??
plt.figure(figsize = (15, 10))
sns.heatmap(data.isnull().T, xticklabels='auto', yticklabels='auto')
plt.show()
```

```python
#get mean and median data for key variables
mean_price = newdf.groupby('city').agg({'price': 'mean'})
mean_reviews = newdf.groupby('city').agg({'reviews_per_month': 'mean'})
mean_availability=newdf.groupby('city').agg({'availability_365': 'mean'})
median_price = newdf.groupby('city').agg({'price': 'median'})
median_reviews = newdf.groupby('city').agg({'reviews_per_month': 'median'})
median_availability=newdf.groupby('city').agg({'availability_365': 'median'})
```

```python
mean_df = pd.DataFrame({
    
    "Mean Price": [mean_price],
    "Median Price": [median_price],
    "Mean Number reviews": [mean_reviews], 
    "Median Number reviews":[median_reviews],
    "Mean nights available": [mean_availability],
    "Median night available": [median_availability] 
 })
mean_df
```

```python
#Visualize frequency of rentals by specific neighborhood
sns.countplot("neighbourhood_group",data=newdf)
plt.xticks(rotation=90)
```

```python
#Visualize frequency of rentals by specific city
sns.countplot("city",data=newdf)
plt.xticks(rotation=90)
```

```python
#visualize room type
newdf['room_type'].value_counts().plot(kind='barh', figsize=(6,4), 
                                     edgecolor=(0,0,0),color='r', title='Type of location')
```

```python
# view descriptive statis for price
newdf['price'].describe()
```

```python
# Create the bins in which Data will be held

bins = [0, 100, 200, 400, np.inf]

# Create the names for the five bins
group_names = ["100 and below", "100 to 200", "200 to 400", "400 or more"]


newdf['price_bins'] = pd.cut(newdf['price'], bins, labels=group_names, include_lowest=True)
newdf
```
